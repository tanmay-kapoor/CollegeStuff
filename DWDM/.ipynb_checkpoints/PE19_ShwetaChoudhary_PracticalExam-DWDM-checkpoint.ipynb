1)Handling missing values :-

```python
#a)Fillna

dataset.isnull().sum()                                                           #Finding sum of all individual null values
```

```python
dataset['Paid'].fillna(0.0,inplace=True)                                    #Filling all categorical values with 0.0
dataset.isnull().sum()
```

```python
#b)dropna
dataset.dropna(subset=['like'])                                                       #dropping null values in like column
dataset.like.isnull().sum()
```

```python
#c)replace 
dataset.fillna({'share':0}, inplace=True)                                             #replacing all null in like with 0

dataset.share.isnull().sum()
```

```python
#d)Data smoothing (BINNING):-

data = np.sort(dataset['Lifetime Post Total Reach'])            #sorting according to column - Lifetime Post Total Reach        
print(data[:250])

print(len(data))
```

```python
n = len(data)

b1=np.zeros((n/5,5))                                                                    #for smoothing by bin mean
b2=np.zeros((n/3,3))                                                                    #for smoothing by bin median
b3=np.zeros((n/3,3))                                                                    #for smoothing by bin boundary
```

```python
#Smoothing by bin Mean 
mean = 0
for i in range (1,n,5): 
  k=int(i/5) 
  mean= (mean + data[i])/5
  for j in range(6): 
    b1[k-1,j-1]=mean 

print("-----------------Smoothing by bin mean-----------------")
print(b1[:10])                                                                          #printing only 1st 10 bins out of total
```

```python
#Smoothing by bin Median
for i in range (1,n,3): 
  k=int(i/3) 
  for j in range (3): 
    b2[k,j]=data[i+1] 
print("-----------------Smoothing by bin median----------------- \n")
print(b2[:10])                                                                          #printing only 1st 10 bins out of total
```

```python
#Smoothing by bin Median
for i in range (1,n,3): 
  k=int(i/3) 
  for j in range (3): 
    if (data[i+j]-data[i]) < (data[i+2]-data[i+j]): 
      b3[k,j]=data[i] 
    else: 
      b3[k,j]=data[i+2]   

print("-----------------Smoothing by bin boundary----------------- \n")
print(b3[:10])                                                                          #printing only 1st 10 bins out of total
```

2)DATA REDUCTION :-

```python
#a) Histogram :- (equi-frequency)

def equalObs(x, nbin):
    nlen = len(x)
    return np.interp(np.linspace(0, nlen, nbin + 1),
                     np.arange(nlen),
                     np.sort(x))

n, bins, patches = plt.hist(dataset['Page total likes'], equalObs(dataset['Page total likes'], 10), edgecolor='black')
plt.show()
```
