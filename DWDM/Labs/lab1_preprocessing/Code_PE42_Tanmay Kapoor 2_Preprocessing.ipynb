# Z-score normalization

Z score = (x -mean) / std. deviation


68% of the data points lie between +/- 1 standard deviation.


95% of the data points lie between +/- 2 standard deviation


99.7% of the data points lie between +/- 3 standard deviation

```python
bills = tips_data["total_bill"]
plt.hist(bills)
plt.show()
```

```python
bills[3] = -268
bills[99] = 424
```

```python
mean = np.mean(bills) 
std = np.std(bills) 
print('mean of the dataset is', mean) 
print('std. deviation is', std) 
```

```python
threshold = 3
outlier = [] 
for i in bills: 
    z = (i-mean)/std 
    if z > threshold or z < -threshold: 
        outlier.append(i) 
print('outlier in dataset is', outlier) 
```

# Sampling


```python
tips_data = sns.load_dataset("tips")
tips_data.head()
tips_data.shape
```

### by default without replacement

```python
tips_data.sample(10)
```

```python
weight_col = [0.0, 0.2, 0.2, 0.4, 0.9, 0.5, 0.2, 0.1, 0.5, 0.7] 
tips_data.head(10).sample(n=5, weights = weight_col)
```

### With Replacement

```python
tips_data.head(10).sample(10, replace = True)
```

```python
weight_col = [0.3, 0.2, 0.2, 0.4, 0.9, 0.5, 0.2, 0.1, 0.5, 0] 
tips_data.head(10).sample(10, replace = True, weights = weight_col)
```
