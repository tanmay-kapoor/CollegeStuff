```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
import pyfpgrowth as fp
import time 
```

```python
orders_data = pd.read_csv("datasets/restaurant-1-orders.csv")
```

```python
orders_data.head()
```

```python
df = orders_data.drop(['Quantity', 'Product Price'], axis = 1)
```

```python
cc = df.groupby(['Order Number', 'Order Date', 'Total products']).cumcount() + 1
cc
```

```python
df = df.set_index(['Order Number', 'Order Date', 'Total products', cc]).unstack().sort_index(1, level=1)
```

```python
df.columns = ['_'.join(map(str,i)) for i in df.columns]
```

```python
final_df = df.reset_index()
final_df
```

```python
final_df.shape
```

```python
start_time = time.time()
records = []
for i in range(20, 1000):
    order = []
    for j in range (3, 10):
        if pd.isna(final_df.iloc[i,j]) == False:
                order.append(final_df.values[i][j])
    records.append(order)

total_time = time.time() - start_time
print(total_time)
#if final_df.values[i][j] is not None:
```

```python
records
```

# APRIORI ALGORITHM

```python
#Building the first apriori model
start_time = time.time()
association_rules = apriori(records, min_support=0.050, min_confidence=0.5, min_lift=1.2, min_length=2)
association_results = list(association_rules)
total_time = time.time() - start_time
print("Time taken in Apriori" + str(total_time))
```

```python
#getting the number of rules
print(len(association_results))
```

```python
#Glancing at the rule
print(association_results)
```

```python
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]) + "\n")
```

# FP GROWTH ALGORITHM

```python
start_time = time.time()
patterns = fp.find_frequent_patterns(records, 100)
association_rules = fp.generate_association_rules(patterns, .30)          #chances of taking are more than 20%
total_time = time.time() - start_time
print(len(patterns))
print(patterns)
print("\n\nrules -> ")
print(association_rules)
print("\nTime taken in FP growth = " + str(total_time))
```

```python
for key,value in association_rules.items():
    print(key[0] + '->' + str(value[0]))
    print(value[1])
    print("\n")
```
