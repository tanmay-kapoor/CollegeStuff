```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
dataset = pd.read_csv('groceries.csv', header = None)
```

```python
dataset.head()
```

```python
update_dataset = dataset.drop([0]) 
```

```python
update_dataset.head()
```

```python
transactions = dataset.iloc[:, :].astype(str).values.tolist()
```

```python
rules = apriori(transactions, min_support = 0.002, min_confidence = 0.8, min_lift = 3, min_length = 2)
```

```python
results = list(rules)
```

```python
listed_rules = []

for i in range(0, len(results)):
    listed_rules.append('Rule: ' + str(list(results[i][0])[0]) +
                        ' -> ' + str(list(results[i][0])[1]) +
                        ' S: ' + str(round(results[i].support, 4)) +
                        ' C: ' + str(round(results[i][2][0].confidence, 4)) +
                        ' L: ' + str(round(results[i][2][0].lift, 4)))
```
